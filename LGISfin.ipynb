```python
def LDS(nums, n):
    dp = []
    dp_list = []

    for i in range(n):
        dp.append(1)
        dp_list.append([nums[i]])
        for j in range(i):
            if nums[j] > nums[i]:
                dp[i] = max(dp[i], dp[j] + 1)
                if len(dp_list[i]) <= len(dp_list[j]):
                    dp_list[i] = dp_list[j] + [nums[i]]

    l2 = dp_list[dp.index(max(dp))]
    print(*l2)
    return dp, dp_list
```
